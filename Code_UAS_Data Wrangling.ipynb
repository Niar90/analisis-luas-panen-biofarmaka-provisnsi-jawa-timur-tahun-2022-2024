```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
df_2022 = pd.read_excel("data_biofarmaka/2022.xlsx")
df_2023 = pd.read_excel("data_biofarmaka/2023.xlsx")
df_2024 = pd.read_excel("data_biofarmaka/2024.xlsx")
```

# Data Cleaning

```python
# Nama kolom dibersihkan dari spasi berlebih, diubah menjadi huruf kecil, dan semua spasi diganti dengan underscore
def clean_columns(df):
    df.columns = df.columns.str.strip()
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(" ", "_")
    return df

# Clean header
df_2022 = clean_columns(df_2022)
df_2023 = clean_columns(df_2023)
df_2024 = clean_columns(df_2024)
```

```python
# Hapus baris kosong
df_2022 = df_2022.dropna(how='all')
df_2023 = df_2023.dropna(how='all')
df_2024 = df_2024.dropna(how='all')
```

```python
# Tambahkan kolom tahun
df_2022["tahun"] = 2022
df_2023["tahun"] = 2023
df_2024["tahun"] = 2024
```

```python
# Samakan urutan kolom
df_2023 = df_2023[df_2022.columns]
df_2024 = df_2024[df_2022.columns]
```

# Data Integration

```python
df_final = pd.concat([df_2022, df_2023, df_2024], ignore_index=True)
df_final.head()
```

```python
# Mengubah nama kolom kabupaten/kota menjadi kabupaten_kota agar lebih rapi dan mudah diolah.
df_final = df_final.rename(columns={
    'kabupaten/kota': 'kabupaten_kota'
})
# Membersihkan data di kolom kabupaten_kota dari spasi berlebih di awal atau akhir teks, supaya datanya konsisten.
df_final['kabupaten_kota'] = df_final['kabupaten_kota'].astype(str).str.strip()
```

```python
# Kata yang ingin dihapus atau dikecualikan dari data.
exclude = ["jawa timur", "total", "jumlah", "provinsi", "catatan", "angka tetap"]

df_final = df_final[~df_final['kabupaten_kota'].str.lower().str.contains('|'.join(exclude))]
```

```python
df_final = df_final.drop_duplicates()
```

# Normalisasi dan Renaming Kolom

```python
# Normalisasi teks untuk kolom kategori
cat_col = df_final.select_dtypes(include="object").columns

for col in cat_col:
    df_final[col] = df_final[col].astype(str).str.title().str.strip()
```

```python
# Ambil semua kolom luas_panen
num_cols = [col for col in df_final.columns if 'luas_panen' in col]
```

```python
# Ubah seluruh kolom luas_panen menjadi numeric
# Kalau ada simbol atau string aneh, otomatis jadi NaN
df_final[num_cols] = df_final[num_cols].apply(
    lambda s: pd.to_numeric(s, errors='coerce')
)
```

```python
# Kode itu dibuat untuk mengambil semua nama kolom yang ada kata 'luas_panen' di dalamnya. Jadi, ini untuk mengumpulkan semua kolom tentang luas panen tanaman.
df_final[num_cols] = df_final[num_cols].fillna(0)
```

```python
# Untuk merapikan index DataFrame setelah diolah
df_final = df_final.reset_index(drop=True)
```

```python
# Renaming kolom
satuan_pohon = ["mahkota_dewa", "mengkudu/pace", "jeruk_nipis"]

new_columns = {}

for col in df_final.columns:
    if "luas_panen" in col:
        nama_tanaman = col.replace("luas_panen_", "")
        nama_tanaman = nama_tanaman.split("_(")[0]

        if nama_tanaman in satuan_pohon:
            satuan = "pohon"
        else:
            satuan = "m²"

        if nama_tanaman == "serai":
            satuan = "m²"

        new_columns[col] = f"luas_panen_{nama_tanaman} ({satuan})"
    else:
        new_columns[col] = col

df_final = df_final.rename(columns=new_columns)
```

# Analisis Perbandingan Antar Tahun

```python
# daftar semua kolom luas panen (kecuali kabupaten & tahun)
panen_cols = df_final.columns.drop(['kabupaten_kota', 'tahun'])

# buat total per baris
df_final['total_luas_panen'] = df_final[panen_cols].sum(axis=1)
df_final.head()
```

```python
# semua kolom luas panen
tanaman_cols = [col for col in df_final.columns if "luas_panen" in col]
tanaman_cols
```

```python
# Total luas panen semua tanaman per tahun
total_per_tahun = df_final.groupby("tahun")[tanaman_cols].sum().sum(axis=1)
total_per_tahun
```

```python
top10 = (
    df_final.groupby(['tahun', 'kabupaten_kota'])['total_luas_panen']
    .sum()
    .sort_values(ascending=False)
    .groupby('tahun')
    .head(10)
    .reset_index()
)

plt.figure(figsize=(10,6))
sns.barplot(
    data=top10,
    x="total_luas_panen",
    y="kabupaten_kota",
    hue="tahun"
)
plt.title("Top 10 Kabupaten/Kota dengan Total Luas Panen Biofarmaka Terbesar")
plt.xlabel("Total Luas Panen")
plt.ylabel("Kabupaten_Kota")
plt.legend(title="Tahun")
plt.show()
```

```python
# pilih 20 kab/kota dengan total panen terbesar
top20 = (
    df_final.groupby("kabupaten_kota")["total_luas_panen"]
    .sum()
    .sort_values(ascending=False)
    .head(20)
    .index
)

pivot_heatmap = df_final[df_final["kabupaten_kota"].isin(top20)].pivot_table(
    index="kabupaten_kota",
    columns="tahun",
    values="total_luas_panen",
    aggfunc="sum"
)

plt.figure(figsize=(10, 8))
sns.heatmap(
    pivot_heatmap,
    cmap="YlGnBu",
    linewidths=.3,
    linecolor="white",
    annot=True,
    fmt=".0f",
    cbar_kws={"shrink": 0.6},
)
plt.title("Heatmap Total Luas Panen Biofarmaka (Top 20 Kabupaten/Kota)", fontsize=14)
plt.xlabel("Tahun")
plt.ylabel("Kabupaten/Kota")
plt.tight_layout()
plt.show()
```

```python
tanaman_cols = [
    col for col in df_final.columns
    if "luas_panen" in col and col not in ["total_luas_panen"]
]

total_per_tanaman = df_final[tanaman_cols].sum().sort_values(ascending=False)

top5_tanaman = total_per_tanaman.head(5).index

df_trend_tanaman = df_final.groupby("tahun")[top5_tanaman].sum()

df_trend_tanaman.index = df_trend_tanaman.index.astype(int)

# Plot
plt.figure(figsize=(10,6))
for col in df_trend_tanaman.columns:
    plt.plot(
        df_trend_tanaman.index,
        df_trend_tanaman[col],
        marker="o",
        linewidth=2,
        label=col
    )

plt.title("Tren 5 Tanaman Biofarmaka Terbesar (2022–2024)")
plt.xlabel("Tahun")
plt.ylabel("Total Luas Panen")
plt.grid(alpha=0.3)
plt.legend()
plt.xticks(df_trend_tanaman.index)
plt.show()
```

```python
# Ambil semua kolom tanaman (selain kategori dan tahun)
plant_cols = [
    col for col in df_final.columns
    if col not in ["kabupaten_kota", "tahun", "total_luas_panen"]
]

# Bersihkan nama tanaman biar enak dibaca
plant_names = [
    col.replace("luas_panen_", "")
       .replace("(meter_persegi)_(m2)", "")
       .replace("(pohon)_(pohon)", "")
       .replace("_", " ")
       .title()
    for col in plant_cols
]

# Buat df_rank
df_rank = (
    df_final.groupby(['tahun', 'kabupaten_kota'])['total_luas_panen']
    .sum()
    .sort_values(ascending=False)
    .groupby('tahun')
    .head(10)
    .reset_index()
)

# Tambahkan kolom berisi daftar tanaman yang dihitung
df_rank["tanaman_dihitung"] = ", ".join(plant_names)

df_rank.head()
```

```python
# Total luas panen per kabupaten berdasarkan tahun
# total luas panen (semua tanaman) per baris
df_final["total_luas"] = df_final[tanaman_cols].sum(axis=1)

# pivot: kabupaten sebagai baris, tahun sebagai kolom
df_pivot = df_final.pivot_table(
    index="kabupaten_kota",
    columns="tahun",
    values="total_luas",
    aggfunc="sum"
).fillna(0)

df_pivot
```

```python
# Selisih antar tahun
tahun_list = sorted(df_final["tahun"].unique())

for i in range(1, len(tahun_list)):
    prev_t = tahun_list[i-1]
    curr_t = tahun_list[i]
    df_pivot[f"selisih_{curr_t}_vs_{prev_t}"] = df_pivot[curr_t] - df_pivot[prev_t]

df_pivot
```

```python
# Persentase perubahan (growth rate)
for i in range(1, len(tahun_list)):
    prev_t = tahun_list[i-1]
    curr_t = tahun_list[i]
    df_pivot[f"persentase_{curr_t}_vs_{prev_t}"] = (
        (df_pivot[curr_t] - df_pivot[prev_t]) / df_pivot[prev_t].replace(0, 1)
    ) * 100

df_pivot
```

```python
# Ranking Perubahan Antar Tahun (Kabupaten yang Paling Naik/Turun)
ranking_perubahan = df_pivot.sort_values("selisih_2023_vs_2022", ascending=False)
ranking_perubahan.head(10)
```

```python
# Deteksi outlier (kabupaten dengan peningkatan tidak wajar)
df_pivot["zscore_2024"] = (
    (df_pivot[2024] - df_pivot[2024].mean()) / df_pivot[2024].std()
)

df_pivot["zscore_2023"] = (
    (df_pivot[2023] - df_pivot[2023].mean()) / df_pivot[2023].std()
)

df_pivot["zscore_2022"] = (
    (df_pivot[2022] - df_pivot[2022].mean()) / df_pivot[2022].std()
)

df_pivot
```
